```python
import pandas as pd
```

```python
data = pd.read_excel(r'/content/gdrive/My Drive/Colab Notebooks/ЛР1 ПАД_Исходные данные_Бухов.xlsx')
Y = data.iloc[:, 0] # множество зависимых переменных (регрессанты)
X = data.iloc[:, 1:] # множество независимых переменных (регрессоры)
#data.head(6)
X
```

```python
import numpy as np
```

```python
# Нормирование данных (Z-нормализация)
mean = np.mean(X, axis=0)
dev = np.std(X, axis=0)
X_norm = (X - mean) / dev

x0 = np.ones((len(data.index), 1))
X_norm.insert(0, "x0", x0)
X_norm.head()
```

```python
# Аналитический метод расчета весов линейной регрессии
w = np.linalg.inv(X_norm.T.dot(X_norm)).dot(X_norm.T).dot(Y)
w
```

```python
# Построение корреляционной матрицы
corr_matr = X_norm.corr()
corr_matr
```

```python
# Число обусловленности
onum = np.linalg.cond(X_norm)
onum
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Считаем среднеквадратичное отклонение
Y_pr = np.matmul(X_norm, w)
MSE = mean_squared_error(Y, Y_pr)
print("MSE: ", MSE)
```

```python
#Анализ регрессионных остатков
from sklearn.metrics import r2_score

#Отклонение
Y_pr = np.matmul(X_norm, w)
MSE = mean_squared_error(Y, Y_pr)
RMSE = MSE ** 0.5
print("RMSE: ", RMSE)

#Коэффициент детерминации
d_coef = r2_score(Y, Y_pr)
print("Det_coef: ", d_coef)
```

```python
# Метод градиентного спуска

w_grad = np.ones(len(data.columns)) # изначально задает единичные веса
learn_rate = 0.1
iter_count = 1500
error_arr = []

for _ in range(iter_count):

  #Функция потерь MSE
  Y_pred = np.matmul(X_norm, w_grad)
  MSE = mean_squared_error(Y, Y_pr)
  error_arr.append(MSE)

  #Градиент ф-ции потерь MSE
  dMSE_dW = 2/(len(data.index)) * (Y - Y_pred).dot(-X_norm)
  w_grad -= learn_rate * dMSE_dW

print(w_grad)
print(error_arr)
```
